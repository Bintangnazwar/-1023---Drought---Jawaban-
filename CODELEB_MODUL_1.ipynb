COBELEB 1

a. Buat 12 tuple (masing-masing 4 data: int, float, string acak)

```python
data_tuple = [
    (1, 2.5, "apel", 10),
    (3.2, 7, "mangga", 5),
    ("jeruk", 8, 4.5, 2),
    (9, "pisang", 6.7, 3),
    (4, 1.1, "melon", 8),
    ("anggur", 2, 3.3, 6),
    (5, "kiwi", 9.9, 1),
    (7.7, 4, "leci", 12),
    (11, "durian", 2.2, 14),
    (13, 6.6, "salak", 15),
    ("nanas", 16, 8.8, 17),
    (18, "pepaya", 3.4, 19)
]
```

b. Konversi tuple ke daftar (list)

```python
data_list = []

for tup in data_tuple:
    for item in tup:
        data_list.append(item)

print(data_list)
```

c. Tambahkan 2 data baru pada indeks sesuai digit terakhir NIM

```python
nim_terakhir = 3

data_list.insert(nim_terakhir, "DATA_BARU_1")
data_list.insert(nim_terakhir + 1, "DATA_BARU_2")

print(data_list)
```

d. Hapus tipe data float dari daftar

```python
data_list = [x for x in data_list if not isinstance(x, float)]

print(data_list)
```

e. Urutkan dalam urutan menurun

```python
data_list = sorted(data_list, key=lambda x: str(x), reverse=True)

print(data_list)
```

CODELEB 2

a. Buat array 2D

```python
import numpy as np

nim = "202310370311063"
# Ambil 6 digit terakhir
last_6 = nim[-6:]
# Ubah ke list integer
nim_digits = [int(x) for x in last_6]
# Buat matriks 2x3
matrix = np.array(nim_digits).reshape(2, 3)

print("Matriks 2D:")
print(matrix)
```

b. Operasi Matriks

```python
# Tambah
print("Penjumlahan:")
print(matrix + matrix)

# Kurang
print("Pengurangan:")
print(matrix - matrix)

# Kali
print("Perkalian:")
print(matrix * matrix)
```

c. Konversi ke 1D

```python
matrix_1d = matrix.flatten()

print(matrix_1d)
```

d. Pengirisian

```python
print("Baris pertama:", matrix[0])
print("Kolom kedua:", matrix[:, 1])
print("Elemen tengah:", matrix[0, 1])
```
